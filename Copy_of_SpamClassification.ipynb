```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
from sklearn import preprocessing, metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import ComplementNB, GaussianNB, MultinomialNB
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from matplotlib import pyplot as plt
from sklearn.preprocessing import MaxAbsScaler
```

```python
#Two models + 5 classifiers.
#Classification of Spam Emails.
#Train Test Split Model with Logistic Regression, multinomialNB, ConplementNB, GaussianNB, and Random Forest Classifiers
#Next was a Deep Nueral Network along with a graph plotting the accuracy of the model.
```

```python
#Load Data
df = pd.read_csv('emails.csv')
#Prep Data - collumn had the string "Email" a space then the number, caused an error.
df.drop('Email No.', axis=1, inplace=True)
df.sample(10)
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Prediction', axis=1), df['Prediction'], test_size=0.33, random_state=42)
scaler = preprocessing.MaxAbsScaler().fit(X_train)
X_scaled = scaler.transform(X_train)
```

```python
print("\nLogisticRegression\n")
clf = LogisticRegression()
clf.fit(X_train, y_train)
#clf.fit(X_scaled, y_train)
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print("Accuracy:", accuracy_score(y_pred, y_test)*100)
print("Recall:", recall_score(y_pred, y_test)*100)
print("Precision:", precision_score(y_pred, y_test)*100)
print("F1 Score:", f1_score(y_pred, y_test)*100)

print("\nMultinomialNB\n")
clf = MultinomialNB()
clf.fit(X_train, y_train)
y_predNB=clf.predict(X_test)
print(confusion_matrix(y_test, y_predNB))
print("Accuracy:", accuracy_score(y_predNB, y_test)*100)
print("Recall:", recall_score(y_predNB, y_test)*100)
print("Precision:", precision_score(y_predNB, y_test)*100)
print("F1 Score:", f1_score(y_predNB, y_test)*100)

print("\nCompliment NB\n")
clf = ComplementNB()
clf.fit(X_train, y_train)
y_predNB=clf.predict(X_test)
print(confusion_matrix(y_test, y_predNB))
print("Accuracy:", accuracy_score(y_predNB, y_test)*100)
print("Recall:", recall_score(y_predNB, y_test)*100)
print("Precision:", precision_score(y_predNB, y_test)*100)
print("F1 Score:", f1_score(y_predNB, y_test)*100)

print("\nGaussian Naive Bayes\n")

clf = GaussianNB()
clf.fit(X_train, y_train)
print(clf.predict(X_test))
y_pred_GB = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred_GB))
print("Accuracy:", accuracy_score(y_pred_GB, y_test)*100)
print("Recall:", recall_score(y_pred_GB, y_test)*100)
print("Precision:", precision_score(y_pred_GB, y_test)*100)
print("F1 Score:", f1_score(y_pred_GB, y_test)*100)

print("\nRandom Forest\n")

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
y_pred_rf = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_pred_rf, y_test)*100)
print("Recall:", recall_score(y_pred_rf, y_test)*100)
print("Precision:", precision_score(y_pred_rf, y_test)*100)
print("F1 Score:", f1_score(y_pred_rf, y_test)*100)

```

```python
pd.options.display.max_rows = 10
pd.options.display.float_format = "{:.1f}".format
```

```python
def create_model(my_learning_rate):
  """Create and compile a deep neural net."""

  model = tf.keras.models.Sequential()
#L1
  model.add(tf.keras.layers.Dense(units=30, activation='relu', name='Hidden1'))
#L2
  model.add(tf.keras.layers.Dense(units=15, activation='relu', name='Hidden2'))
#Dropout
  model.add(tf.keras.layers.Dropout(rate=0.2))
#Output
  model.add(tf.keras.layers.Dense(units=2, activation='softmax'))

  model.compile(optimizer=tf.keras.optimizers.Adam(lr=my_learning_rate),
                loss="sparse_categorical_crossentropy",
                metrics=['accuracy'])

  return model
```

```python
def plot_curve(epochs, hist, list_of_metrics):
  """Plot a curve of one or more classification metrics vs. epoch."""

  plt.figure()
  plt.xlabel("Epoch")
  plt.ylabel("Value")

  for m in list_of_metrics:
    x = hist[m]
    plt.plot(epochs[1:], x[1:], label=m)

  plt.legend()

print("Loaded the plot_curve function.")
```

```python
def train_model(validation_split, model, train_features, train_label, epochs,
                batch_size=None):
  """Train the model by feeding it data."""

  history = model.fit(x=train_features, y=train_label, batch_size=batch_size,
                      epochs=epochs, shuffle=True,
                      validation_split=validation_split)
  epochs = history.epoch
  hist = pd.DataFrame(history.history)

  return epochs, hist

# hyperparameters
learning_rate = 1
epochs = 70
batch_size = 2000
validation_split = 0.02

my_model = create_model(learning_rate)

# Train the model on set.
epochs, hist = train_model(validation_split, my_model, X_scaled, y_train,
                           epochs, batch_size)
list_of_metrics_to_plot = ['accuracy']
plot_curve(epochs, hist, list_of_metrics_to_plot)
```
